```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
# Data Agrregation consists of operations that result
# in a scalar (e.g. mean(),sum(),count(), etc)

#Let's get a csv data set to play with
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
```

```python
# From the web site, download the csv file to same python notebook folder
# Save thewinequality.csv file in the same folder
# as your ipython notebooks, note the delimiter used ;
dframe_wine = pd.read_csv('winequality_red.csv',sep=';')
```

```python
# Let's get a preview
dframe_wine.head()
```

```python
# How about we find out the average alcohol content for the wine
dframe_wine['alcohol'].mean()
```

```python
# That was an example of an aggregate, how about we make our own?
# Instead of calling .mean, .sum or .count etc
# Create a function using def
# Call the function max_to_min
# Find the max and subtract min
def max_to_min(arr):
    return arr.max() - arr.min()

# Let's group the wines by "quality"
wino = dframe_wine.groupby('quality')

# Show stats on that wine (by quality)
wino.describe()
```

```python
# We can now apply our own aggregate function, this function takes the max value of the col
# and subtracts the min value of the col
# Call the method agg and pass the function to it
wino.agg(max_to_min)
```

```python
# We can also pass string methods through aggregate
# Can pass string methods too agg
# Pass mean as a string
# Instead of using .mean
wino.agg('mean')
```

```python
# Let's go back to the original dframe
dframe_wine.head()
```

```python
# Let's adda  quality to alcohol content ratio
# Added to last column
dframe_wine['qual/alc ratio'] = dframe_wine['quality']/dframe_wine['alcohol']
```
