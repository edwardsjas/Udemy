```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#####Scatter Plots

```python
# Now create a scatter plot of test scores vs time spent
# Look for a trend or a relationship between the two
test_scores = [80,68,52,81,64,90,48,58,62,72,54,88]
time_spent = [28,29,32,40,26,34,38,44,22,30,40,24]
plt.scatter(time_spent, test_scores)

plt.title('Test Score vs Time Spent')
plt.xlabel('Time Spent on Test')
plt.ylabel('Test Score')
plt.show()

x = [1,2,3,4,5]
y1 = [2,3,2,4,2]
y2 = [8,8,6,7,6]

plt.scatter(x,y1, marker='o', color='c')
plt.scatter(x,y2, marker='v', color='m')
plt.title('Scatter Graph 2')
```
