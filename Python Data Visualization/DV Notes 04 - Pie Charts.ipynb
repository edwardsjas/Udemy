###Pie Charts
* x is amount/values
* y is values
* Can pass colors

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a pie chart
# Assign labels, sizes (values) and colors
labels = 'Taxes','Overheads','Entertainment'
sizes = [25,32,12]
colors = ['c','m','b']

# Use plt.pie
# Pass sizes as x (amount/values)
# Pass labels and colors list
# Use start angle is 90
# Show shadow
# Explode pull out the pieces)
# Show %
plt.pie(sizes,
        labels=labels,
        colors=colors,
        startangle=90,
        shadow=True,
        explode=(0.1,0.1,0.1),
        autopct = '%1.1f%%')
plt.show
```
